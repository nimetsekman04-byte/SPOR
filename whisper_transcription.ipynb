```python
from google.colab import files

uploaded = files.upload()
audio_path = next(iter(uploaded))
print("Yüklenen dosya:", audio_path)
```

```python
import whisper
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

print("Large model yükleniyor...")
model_large = whisper.load_model("large").to(device)

print("Model yüklendi.")
```

```python
print("Deşifre başlıyor...")

result_large = model_large.transcribe(
    audio_path,
    language="tr"
)

print("\n--- LARGE MODEL ÇIKTISI ---\n")
print(result_large["text"])
```

```python
large_filename = "transcript_large.txt"

with open(large_filename, "w", encoding="utf-8") as f:
    f.write(result_large["text"])

print("Dosya kaydedildi:", large_filename)
```
